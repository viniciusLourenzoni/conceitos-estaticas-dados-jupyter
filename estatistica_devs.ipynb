```python
import pandas as pd
```

```python
pd.__version__
```

```python
dict_medidas = {'idade': [15, 18, 25, 27, 29, 31, 33, 35, 37, 39, 41], 'altura': [155, 165, 175, 185, 195, 205, 215, 225, 235, 245, 255]}
```

```python
dict_medidas
```

```python
df_medidas = pd.DataFrame.from_dict(dict_medidas)
```

```python
df_medidas
```

## Medidas de Posição

```python
# media -> A média é a soma de todos os valores dividida pelo número de valores
df_medidas["idade"].mean()
```

```python
# mediana - > A mediana é o valor que separa a metade superior da metade inferior de um conjunto de dados. 
df_medidas["idade"].median()
```

```python
# Moda -> O valor mais frequente em um conjunto de dados
df_medidas["idade"].mode()
```

## Medidas de Disperção

```python
# variancia -> A variância mede a dispersão dos dados em relação à média.
df_medidas.idade.var()
```

```python
# desvio padrão -> O desvio padrão é a raiz quadrada da variância.
df_medidas.idade.std()
```

```python
# coeficiente de variacao -> O coeficiente de variação (CV) é uma medida relativa da dispersão dos dados em relação à média.
df_medidas.idade.std() / df_medidas.idade.mean() * 100
```

## Medidas de Forma

### Assimetria



```python
# Idade
result: float = df_medidas.idade.skew()
print(result)
```

```python
# Altura
result: float = df_medidas.altura.skew()
print(result)
```

### Curtose



```python
# Idade
df_medidas.idade.kurtosis()
```

```python
#Altura
df_medidas.altura.kurtosis()
```

```python
# Obter medidas estatisticas com metodo nativo do pandas
df_medidas.idade.describe()
```

### Correlação de variáveis

```python
# Correlação de Person - Dataframe
df_medidas.corr(method="pearson")
```

```python
# Correlação entre 2 variáveis
df_medidas.idade.corr(df_medidas.altura)
```

```python
# Correlação de Spearman - Dataframe
df_medidas.corr(method="spearman")
```

### Representações Gráficas

```python
# Histograma - Idade
df_medidas.idade.hist()
```

```python
# Histograma- Altura
df_medidas.altura.hist()
```

### Dataframe de Vendas

```python
dict_vendas = {'categoria': ['Masculino', 'Feminino', 'Infantil', 'Casa'], 'valor': [400000, 600000, 250000, 580000], 'quantidade': [3000, 5000, 1500, 2500]}
```

```python
dict_vendas
```

```python
df_vendas = pd.DataFrame.from_dict(dict_vendas)
```

```python
df_vendas
```

```python
# Gráfico de Barras Vertical
df_vendas.plot.bar(x='categoria', y='valor')
```

```python
# Gráfico de Barras Vertical - Valor (ordenado)
df_vendas.sort_values('valor', ascending=False).plot.bar(x='categoria', y='valor')
```

```python
# Gráfico de barras horizontal - Quantidade
df_vendas.plot.barh(x='categoria', y='quantidade')
```

```python
# Gráfico de barras horizontal - Quantidade (ordenado - Ranking)
df_vendas.sort_values('quantidade').plot.barh(x='categoria', y='quantidade')
```

```python
# Gráfico de Dispersão
df_medidas.plot.scatter(x='idade', y='altura')
```

```python
# Gráfico de BoxPlot(Caixa)
df_medidas.idade.plot.box()
```

```python
df_medidas.altura.plot.box()
```
